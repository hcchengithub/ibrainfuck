作为一个用这个语言写过不少有趣东西的程（shen）序（jing）员（bing）我们来粗略地理解一下。首先谈到这个语言的定义和运行原理该语言定义在这样一个环境之上：你有一列无限长的小火车，每个车厢里装了一个数字，初始为0。还有一个列车员，初始在最头上那节车厢上。好了，你把你写的BrainFK程序交给列车员，列车员会做如下的事情：

1. 从左向右、由上自下一个字符一个字符地读取你的程序
2. 当读到`+`的时候，将所在车厢里的数字加一
3. 当读到`-`的时候，将所在的车厢里的数字减一
4. 当读到`>`的时候，跑到后一个车厢去
5. 当读到`<`的时候，跑到前一个车厢去
6. 当读到`[`的时候，如果该车厢里面的数字为0，则跳去执行下一个`]`之后的程序内容
7. 当读到`]`的时候，如果该车想里面的数字不为0，则跳去执行上一个`[`之后的程序内容
8. 当读到`.`的时候，将所在车厢里面的数字翻译成ASCII字符，显示在你的屏幕上
9. 当读到`,`的时候，从等待使用者输入一个ASCII字符，转码成数字写进所在车厢里

接下来让我们仔细看这段程序：





```brainfuck
++++++++++[>+++++++>++++++++++>+++>+<<<<-]>++.>+.+++++++
..+++.>++.<<+++++++++++++++.>.+++.------.--------.>+..>.
```
